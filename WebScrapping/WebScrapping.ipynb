```python
s_old = '080800000'
for i in range(45545,99999):
    s = '0808' + str(i)
    print(s)
    if i%500==0:
        doctors.to_csv("doctors.csv") 

    try:
        num = browser.find_element_by_xpath('//*[@id="numeroColegiado"]')
        num.clear()
        num.send_keys(s)
        browser.implicitly_wait(5)
        
        codigo = browser.find_element_by_xpath('//*[@id="codigoCaptcha"]')
        codigo.clear()
        codigo.send_keys('k28b8')
        browser.implicitly_wait(5)

        buscar = browser.find_element_by_xpath('//*[@id="busquedaPublicaColegiadoBuscar"]')
        buscar.click()

        browser.implicitly_wait(5)

        element = browser.find_element_by_xpath('//*[@id="resultadoBusquedaColegiados"]')
        err1 = 'No ha especificado criterios suficientes de búsqueda.Tiene que buscar por "nombre" y "primer apellido" o por "número de colegiado".'
        err2 = 'No hay datos'
        print(element.text)
        if(element.text !=err1 and element.text!=err2):
                numcoleg = browser.find_element_by_xpath('//*[@id="resultadoBusquedaColegiados"]/table/tbody/tr[1]/td[1]')
            #    print(numcoleg.text)
                fullname = browser.find_element_by_xpath('//*[@id="resultadoBusquedaColegiados"]/table/tbody/tr[1]/td[2]')
            #    print(fullname.text)

                document={'FullName':str(fullname.text), 'NumColegio':str(numcoleg.text)}
                new_doctor =  pd.DataFrame([[str(fullname.text),str(numcoleg.text)]])
                new_doctor.columns = ['Full Name', 'Numero de Colegiado']
                doctors = pd.concat([doctors,new_doctor], axis=0)

                collection.insert_one(document)
        else: 
            print('No hay datos')
        browser.implicitly_wait(5)
    except:
        print('Error')
doctors.drop_duplicates(inplace=True)
doctors.to_csv("doctors.csv")
```
