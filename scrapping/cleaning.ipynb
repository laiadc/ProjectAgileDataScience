```python
import pandas as pd

df = pd.read_csv("scrapped_data.csv", encoding="ansi", index_col="id")


def clean(df):
    to_del_list = [
        "Prof\. dr hab\. med\. ",
        "lek\. med\. lek\. stom\. ",
        "lek\. ",
        "dent\. ",
        "dr n\. med\. ",
        "dr hab\. n\. med\. ",
        "prof\. ",
        "dr n\. o zdr\. ",
        "mgr ",
        "wet\. ",
        "M\. ",
        "A\. ",
        "z d\. ",
        "d\. ",
        "nadzw\. ",
        "A\. ",
        "W\. ",
        "Z\. ",
        "W\. ",
        "D\. ",
        "n\. ",
        "K\. ",
        "J\. ",
        " D.O.",
        "K\. ",
        "K\. ",
        "inż\. ",
        "dr hab\. ",
        "dr ",
        "B\. ",
        "Test",
    ]
    #     remove parts of names
    for to_del in to_del_list:
        df["name"] = df["name"].str.replace(to_del, "")

    #     remove whole rows
    to_remove_list = [
        "Dr. Wiankowska",
        "Dr. Olejniczakowska",
        "Krzysztof Grzegorz Szymański vel Szymanek",
        "Krzysztof Szymański vel Szymanek",
        "SC2Labs Penetrationing-TGG99",
    ]
    for to_remove in to_remove_list:
        df = df[df.name != to_remove]

    #     remove only additonal dots
    df["name"] = df["name"].str.replace(".", "")

    #     change parenthesis to dash
    df["name"] = df["name"].str.replace(" \(", "-")
    df["name"] = df["name"].str.replace("\)", "")

    #     remove spaces around dashes
    df["name"] = df["name"].str.replace("–", "-")
    df["name"] = df["name"].str.replace(" - ", "-")

    #     remove triple and double spaces
    df["name"] = df["name"].str.replace("   ", " ")
    df["name"] = df["name"].str.replace("  ", " ")

    #     drop rows with same name
    df = df.drop_duplicates(subset="name")
    return df


df = clean(df)
df.to_csv('scrapped_data.csv', index='id')
```
